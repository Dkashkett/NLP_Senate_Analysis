```python
import pandas as pd
```

```python
dem = pd.read_csv('./dem_results.csv').drop(['Unnamed: 0'], axis=1)
dem.columns = ['bigram','count']
rep = pd.read_csv('./rep_results.csv').drop(['Unnamed: 0'], axis=1)
rep.columns = ['bigram','count']
dem = dem.sort_values(by='count', ascending=False).head(200).reset_index().drop(['index'], axis=1)
rep = rep.sort_values(by='count', ascending=False).head(200).reset_index().drop(['index'], axis=1)
```

```python
inner = pd.merge(dem,rep,on='bigram',how='inner')
inner.columns = ['bigram','dem','rep']
outer = pd.merge(dem,rep,on='bigram',how='outer')
outer.columns = ['bigram','dem','rep']
dem_bigrams = set(dem.bigram)
rep_bigrams = set(rep.bigram)
```

```python
inner['dem_perc'] = (inner.dem -inner.rep)/inner.rep
```

```python
inner['rep_perc'] = (inner.rep-inner.dem)/inner.dem
```

```python
inner.drop(['rep_perc'], axis=1).sort_values('dem_perc', ascending=False).head(12)
```
